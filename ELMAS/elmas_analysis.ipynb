#### Data Components:
The ELMAS dataset provides hourly load profiles representative of industrial and tertiary sectors in France, aggregated into 18 clusters using a K-means algorithm with manual reclassification for interpretability.

#### Profile Volume:  
- **Time Resolution:** 60 minutes  
- **Points per Day:** 24  
- **Total Duration Covered:** 365 days (non-leap year)  
- **Clustering Method:** K-means with manual reclassification by NACE distribution  
> The dataset is based on 55,730 anonymized time series grouped by NACE codes and subscribed power levels.

---

#### Description:  
The ELMAS dataset offers 18 representative load profiles derived from detailed French industrial and tertiary electricity consumption data in 2018. Load patterns are categorized using the NACE classification and grouped by similarity into clusters. Manual reclassification enhances sector interpretability while maintaining statistical integrity. The dataset supports time-series modeling, demand forecasting, and energy system analysis at the national level.

> **Note**: All data is based on real consumption; the dataset is not synthetic.  
> Units:  
> - **Load**: kWh  
> - **Temperature**: °C
