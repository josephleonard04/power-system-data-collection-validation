#### Data Components:  
This dataset contains high-resolution electric load profiles for 50 industrial plants in Germany. The data is split by year and includes two files:

| Profile File               | Description                                                                  | Type        | Role     |
|---------------------------|------------------------------------------------------------------------------|-------------|----------|
| `LoadProfile_20IPs_2016` (20 profiles) | Load profiles of 20 industrial plants for the year 2016                      | Time series | Consumer |
| `LoadProfile_30IPs_2017` (30 profiles)  | Load profiles of 30 different industrial plants for the year 2017            | Time series | Consumer |

#### Profile Volume:
- **Time Resolution:** 15 minutes  
- **Points per Day:** 96 (15-minute intervals × 24 hours)  
- **Total Length per Profile:** 35,040  
- **Duration Covered:** 1 year 
- **Profiles are separate across 2016 and 2017 datasets**

> **Note:** Each profile represents a distinct industrial plant. No site appears in both years.

---

#### Description:  
This dataset, published on [Zenodo](https://doi.org/10.5281/zenodo.3899018), contains anonymized electric load profiles of small and mid-size manufacturing enterprises in Germany. The measurements are recorded at 15-minute intervals and averaged to reflect real operational demand.

It has been used in multiple peer-reviewed publications focusing on industrial energy flexibility, storage optimization, and CO₂-minimization strategies.

> **Units:**  
> - **Load values**: **kW** (average power per 15-minute interval)
