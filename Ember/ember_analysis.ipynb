#### Data Components:
The Ember dataset provides **monthly electricity data** across European countries, covering generation, demand, and emissions. It includes breakdowns by fuel source (coal, gas, nuclear, wind, solar, hydro, etc.), allowing for comparisons across time and geography.

> **Note:** Data is based on official national statistics and third-party sources such as Eurostat and ENTSO-E. All data is aggregated at the national level.

| Profile         | Description                                                   | Type         | Role     |
|-----------------|---------------------------------------------------------------|--------------|----------|
| `Demand`        | National electricity consumption (active power, in TWh)       | Time series  | Consumer |
| `Generation`    | Monthly generation by fuel type (e.g., solar, wind, fossil)   | Time series  | Producer |
| `Emissions`     | CO₂ emissions and intensity (gCO₂/kWh)                        | Time series  | Environment |

- **Time Resolution:** Monthly

- **Total Duration Covered:** Up to 11 years (2015–2025)  
- **Geographies:** Over 30 European countries  
> **Note:** Data coverage varies by country and year, but most EU countries have continuous data from 2015 onward.

---

#### Description:  
The Ember Monthly Electricity Data provides open, curated electricity data at the national level for European countries. It includes demand (load), production (by fuel type), and environmental metrics such as emissions and CO₂ intensity. The dataset is updated twice per month and used by researchers, analysts, and policymakers to track the energy transition in Europe.

> **Note**: Data is real and based on reported national statistics or compiled datasets.  
> **Units:**  
> - **Electricity demand and generation:** **TWh**  
> - **Emissions:** **gCO₂/kWh**, **MtCO₂**
